### Simple random data is as its title suggests - simple and random nature - one doesn't know exactly what number/character is next to be generated. 
### Permutations are by definition a combination of the elements of a list given whch have their elements moved ### around to give a new list but it is really the same list in some other order unless a second parameter is
### passed to the function which can allow the returned list to be of different length to the original list supplied. [#5]

###

```python
### Code for item 3.
### import the functions for graphic output
import matplotlib.pyplot as plt

### import the functions needed to demonstrate the ability of numpy
import numpy as np


# setup the same value for the whole program
n = 1000
```

```python
# To demostrate the RANDN function 

# identify the function type
d="Random"

# clear the list/arrays
x = []
y = []

# generate a series of x values along the X-axis from 0 to 10
x = np.arange(0, n, .1 )

# [#4# generate a random set of y values for each X value
y = np.random.randn(len(x))

# find the max and min for the X values
xmax = max(x)
xmin = min(x)

# find the max and min for the Y values
ymax = max(y)
ymin = min(y)

plt.title("NUMPY "+d+" values")
plt.xlabel('X')
plt.ylabel('Y')

# draw a histogram of the values
plt.hist(y)
# display the graph to the screen
plt.show()
print("Histogram of Distribution:",d)

# create another graph
# give the plot some room on the plot space
plt.xlim(xmin-1,xmax+1),
plt.ylim(ymin-1,ymax+1)
# draw graph of each x and y value on a scatter
plt.scatter(x,y)

# prepare to draw the indication curve
# sort the Y values
y.sort()
# draw graph of each x and y value on a scatter
plt.scatter(x,y)
# display the graph to the screen
plt.show()
print("ScatterPlot of Distribution:",d)

# comment on graph
print("Distribution =",d)
print("Sample size n=",n)
print("BLUE DATA  : Random dots distributed about the graph space.")
print("ORANGE DATA: Slight curling at the outer edges suggests a 'bell curve'")
```

```python
# identify the distribution type
d="Pareto"

# clear the list/arrays
x = []
y = []

# generate a series of x values along the X-axis from 0 to 1000 in steps of .1
x = np.arange(0, n, .1 )

# [#4]
# generate a random set of y values for each X value
#     numpy.random.pareto(a, size=None)
#     Draw samples from a Pareto II or Lomax distribution with specified shape.
y = (np.random.pareto(7,len(x))+1) * 2

# find the max and min for the X values
xmax = max(x)
xmin = min(x)

# find the max and min for the Y values
ymax = max(y)
ymin = min(y)

plt.title("NUMPY "+d+" values")
plt.xlabel('X')
plt.ylabel('Y')

# draw a histogram of the values
plt.hist(y)
# display the graph to the screen
plt.show()
print("Histogram of Distribution:",d)

# create another graph
# give the plot some room on the plot space
plt.xlim(xmin-1,xmax+1),
plt.ylim(ymin-1,ymax+1)

# draw graph of each x and y value on a scatter
plt.scatter(x,y)

# prepare to draw the indication curve
# sort the Y values
y.sort()
# draw graph of each x and y value on a scatter
plt.scatter(x,y)
# display the graph to the screen
plt.show()
print("ScatterPlot of Distribution:",d)

# comment on graph
print("Distribution =",d)
print("Sample size n=",n)
print("Blue   : Biased/Skewed to the left")
print("Orange : No curling on one end suggests skewed and certainly not 'normal'.")
```

```python
# To demostrate the NORMAL function 

# identify the distribution
d="Normal"

# clear the list/arrays
x = []
y = []

# generate a series of x values along the X-axis from 0 to 100 in steps of .1
x = np.arange(0, n, .1 )

# generate a random set of y values for each X value
# [#4] used online numpy documentation for the required input to the Normal function
#           numpy.random.normal(loc=0.0, scale=1.0, size=None)
#                Draw random samples from a normal (Gaussian) distribution.
    
y = np.random.normal(0, .1, len(x))

# find the max and min for the X values
xmax = max(x)
xmin = min(x)

# find the max and min for the Y values
ymax = max(y)
ymin = min(y)

plt.title("NUMPY "+d+" values")
plt.xlabel('X')
plt.ylabel('Y')

# draw a histogram of the values
plt.hist(y)
# display the graph to the screen
plt.show()
print("Histogram of Distribution:",d)

# create another graph
# give the plot some room on the plot space
plt.xlim(xmin-1,xmax+1),
plt.ylim(ymin-1,ymax+1)
# draw graph of each x and y value on a scatter
plt.scatter(x,y)

# prepare to draw the indication curve
# sort the Y values
y.sort()
# draw graph of each x and y value on a scatter
plt.scatter(x,y)
# display the graph to the screen
plt.show()
print("ScatterPlot of Distribution:",d)

# comment on graph
print("Distribution =",d)
print("Sample size n=",n)
print("Blue   : Random but in a band, limited in value.")
print("Orange : Slight curling at the outer edges suggests a 'bell curve'")
```

```python
# To demostrate the ChiSquared function 

# idenitfy the distribution type
d="ChiSquared"

# clear the list/arrays
x = []
y = []

# generate a series of x values along the X-axis from 0 to 100 in steps of .1
x = np.arange(0, n, .1 )

# generate a random set of y values for each X value
# [#4] used online numpy documentation for the required input to the ChiSquared function
#           numpy.random.chisquare(df, size=None)
#               Draw samples from a chi-square distribution.
    
# input value is the Degrees of Freedom which is usually the number of categories minus 1    
y = np.random.chisquare(len(x)-1, len(x))

# find the max and min for the X values
xmax = max(x)
xmin = min(x)

# find the max and min for the Y values
ymax = max(y)
ymin = min(y)

plt.title("NUMPY "+d+" values")
plt.xlabel('X')
plt.ylabel('Y')
# draw a histogram of the values
plt.hist(y)
# display the graph to the screen
plt.show()
print("Histogram of Distribution:",d)

# create another graph
# give the plot some room on the plot space
plt.xlim(xmin-1,xmax+1),
plt.ylim(ymin-1,ymax+1)
# draw graph of each x and y value on a scatter
plt.scatter(x,y)

# prepare to draw the indication curve
# sort the Y values
y.sort()
# draw graph of each x and y value on a scatter
plt.scatter(x,y)
# display the graph to the screen
plt.show()
print("ScatterPlot of Distribution:",d)

# comment on graph
print("Distribution =",d)
print("Sample size n=",n)
print("Blue   : Random scattered.")
print("Orange : Obvious curling at the outer edges suggests a 'bell curve' ")
```

```python
# To demostrate the Uniform function 
d="Uniform"

# clear the list/arrays
x = []
y = []

# [#4] used online numpy documentation for the required input to the ChiSquared function
#        numpy.random.uniform(low=0.0, high=1.0, size=None)
#            Draw samples from a uniform distribution.
# generate a series of x values along the X-axis from 0 to 100 in steps of .1
x = np.arange(0, n, 1 )

# generate a random set of y values for each X value
y = np.random.uniform(min(x), max(x), len(x))

# find the max and min for the X values
xmax = max(x)
xmin = min(x)

# find the max and min for the Y values
ymax = max(y)
ymin = min(y)

plt.title("NUMPY "+d+" values")
plt.xlabel('X')
plt.ylabel('Y')

# draw a histogram of the values
plt.hist(y)
# display the graph to the screen
plt.show()
print("Histogram of Distribution:",d)

# create another graph
# give the plot some room on the plot space
plt.xlim(xmin-1,xmax+1),
plt.ylim(ymin-1,ymax+1)
# draw graph of each x and y value on a scatter
plt.scatter(x,y)

# prepare to draw the indication curve
# sort the Y values
y.sort()
# draw graph of each x and y value on a scatter
plt.scatter(x,y)
# display the graph to the screen
plt.show()
print("ScatterPlot of Distribution:",d)

# comment on graph
print("Distribution =",d)
print("Sample size n=",n)
print("Blue   : Random dots about the graph.")
print("Orange : Straight but slightly rough line suggests all dots equally possible.")
```

##  Explain the use of seeds in generating pseudorandom numbers.

### The use of a seed in generating a so-called random number is to increase the appearance 
### of the number generated being random.  They are pseudo-rando because the generator itself 
### being based on a machine+program must be deterministic and therefore in the true sense of
### the word the output cannot be truly random [#3]
### When a seed is used, it allows the random number generator to start off from a different place 
### similar to someone taking the first step either by left foot or right but determined by a glance 
### at a watch and deciding whether the second is odd or even – the number of outcomes is limited 
### but there is a tint of ‘chance’ in the output.
### If the seed is known and the mehtod to generate the 'random' number is known then the whole 
### step to generate a random number back to its origin can be discovered. [#8]


```python
### Code for item 4.

### example of the code needed to 'seed' the random function
### This code sets the pseudo-random generator off on a different pattern/step than if it had a seed of any other value.
np.random.seed(1234)
```
