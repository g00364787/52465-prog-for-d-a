```python
### Code for item 3.
### import the functions for graphic output
import matplotlib.pyplot as plt

### import the functions needed to demonstrate the ability of numpy
import numpy as np


# setup the same value for the whole program
n = 1000
```

```python
# To demostrate the RANDN function 

# identify the function type
d="Random"

# generate a series of x values along the X-axis from 0 to 10
x = np.arange(0, n, .1 )

# generate a random set of y values for each X value
y = np.random.randn(len(x))

# find the max and min for the X values
xmax = max(x)
xmin = min(x)

# find the max and min for the Y values
ymax = max(y)
ymin = min(y)

plt.title("NUMPY "+d+" values")
plt.xlabel('X')
plt.ylabel('Y')

# give the plot some room on the plot space
plt.xlim(xmin-1,xmax+1),
plt.ylim(ymin-1,ymax+1)

# draw graph of each x and y value on a scatter
plt.scatter(x,y)

# sort the Y values
y.sort()

# draw graph of each x and y value on a scatter
plt.scatter(x,y)

# display the graph to the screen
plt.show()

# comment on graph
print("Distribution =",d)
print("Sample size n=",n)
print("BLUE DATA  : Random dots distributed about the graph space.")
print("ORANGE DATA: Slight curling at the outer edges suggests a wide and nearly flat 'bell curve'")
```

```python
# To demostrate the F function 

# identifythe distribution type
d="F"

# generate a series of x values along the X-axis from 0 to 1000 in steps of 1
x = np.arange(0, n, .1 )

# generate a random set of y values for each X value
# [#4] used online numpy documentation for the required input to the F() function
#             numpy.random.f(dfnum, dfden, size=None)
#                   Draw samples from an F distribution.
y = np.random.f(len(x)-1, len(x)-1, len(x))

# find the max and min for the X values
xmax = max(x)
xmin = min(x)

# find the max and min for the Y values
ymax = max(y)
ymin = min(y)

plt.title("NUMPY "+d+" values")
plt.xlabel('X')
plt.ylabel('Y')

# give the plot some room on the plot space
plt.xlim(xmin-1,xmax+1),
plt.ylim(ymin-1,ymax+1)

# draw graph of each x and y value on a scatter
plt.scatter(x,y)

# sort the Y values
y.sort()

# draw graph of each x and y value on a scatter
plt.scatter(x,y)

# display the graph to the screen
plt.show()

# comment on graph
print("Distribution =",d)
print("Sample size n=",n)
print("Blue   : Random but in a band, limited in value.")
print("Orange : Slight curling at the outer edges suggests a wide and nearly flat 'bell curve'")
```

```python
# To demostrate the NORMAL function 

# identify the distribution
d="Normal"

# generate a series of x values along the X-axis from 0 to 100 in steps of .1
x = np.arange(0, n, .1 )

# generate a random set of y values for each X value
# [#5] used online numpy documentation for the required input to the Normal function
#           numpy.random.normal(loc=0.0, scale=1.0, size=None)
#                Draw random samples from a normal (Gaussian) distribution.
    
y = np.random.normal(0, .1, len(x))

# find the max and min for the X values
xmax = max(x)
xmin = min(x)

# find the max and min for the Y values
ymax = max(y)
ymin = min(y)

plt.title("NUMPY "+d+" values")
plt.xlabel('X')
plt.ylabel('Y')

# give the plot some room on the plot space
plt.xlim(xmin-1,xmax+1),
plt.ylim(ymin-1,ymax+1)

# draw graph of each x and y value on a scatter
plt.scatter(x,y)

# sort the Y values
y.sort()

# draw graph of each x and y value on a scatter
plt.scatter(x,y)

# display the graph to the screen
plt.show()

# comment on graph
print("Distribution =",d)
print("Sample size n=",n)
print("Blue   : Random but in a band, limited in value.")
print("Orange : Slight curling at the outer edges suggests a wide and nearly flat 'bell curve'")
```

```python
# To demostrate the ChiSquared function 

# idenitfy the distribution type
d="ChiSquared"

# generate a series of x values along the X-axis from 0 to 100 in steps of .1
x = np.arange(0, n, .1 )

# generate a random set of y values for each X value
# [#6] used online numpy documentation for the required input to the ChiSquared function
#           numpy.random.chisquare(df, size=None)
#               Draw samples from a chi-square distribution.
    
# input value is the Degrees of Freedom which is usually the number of samples minus 1    
y = np.random.chisquare(len(x)-1, len(x))

# find the max and min for the X values
xmax = max(x)
xmin = min(x)

# find the max and min for the Y values
ymax = max(y)
ymin = min(y)

plt.title("NUMPY "+d+" values")
plt.xlabel('X')
plt.ylabel('Y')

# give the plot some room on the plot space
plt.xlim(xmin-1,xmax+1),
plt.ylim(ymin-1,ymax+1)

# draw graph of each x and y value on a scatter
plt.scatter(x,y)

# sort the Y values
y.sort()

# draw graph of each x and y value on a scatter
plt.scatter(x,y)

# display the graph to the screen
plt.show()

# comment on graph
print("Distribution =",d)
print("Sample size n=",n)
print("Blue   : Random scattered.")
print("Orange : Obvious curling at the outer edges suggests a 'bell curve' ")
```

```python
# To demostrate the Uniform function 
d="Uniform"

# [#7] used online numpy documentation for the required input to the ChiSquared function
#        numpy.random.uniform(low=0.0, high=1.0, size=None)
#            Draw samples from a uniform distribution.
# generate a series of x values along the X-axis from 0 to 100 in steps of .1
x = np.arange(0, n, 1 )

# generate a random set of y values for each X value
y = np.random.uniform(min(x), max(x), len(x))

# find the max and min for the X values
xmax = max(x)
xmin = min(x)

# find the max and min for the Y values
ymax = max(y)
ymin = min(y)

plt.title("NUMPY "+d+" values")
plt.xlabel('X')
plt.ylabel('Y')

# give the plot some room on the plot space
plt.xlim(xmin-1,xmax+1),
plt.ylim(ymin-1,ymax+1)

# draw graph of each x and y value on a scatter
plt.scatter(x,y)

# sort the Y values
y.sort()

# draw graph of each x and y value on a scatter
plt.scatter(x,y)

# display the graph to the screen
plt.show()

# comment on graph
print("Distribution =",d)
print("Sample size n=",n)
print("Blue   : Random dots about the graph.")
print("Orange : Straight & slightly rough line suggests all dots equally possible.")
```
